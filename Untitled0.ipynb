```python
import pandas as panda
import seaborn as sns

#chatGPT recommended using Matlib
import matplotlib.pyplot as plt

from google.colab import files
# uploaded = files.upload()

# Mount Google Drive (if your file is stored in Google Drive)
# from google.colab import drive
# drive.mount('/content/drive')

# Read the CSV file into a Pandas DataFrame
data = panda.read_csv('/content/Video_Games_Sales_as_at_22_Dec_2016.csv')


# Display the first few rows of the DataFrame
print("-----------------------------First 5 records---------------------------")
print(data.head())

# Display all columns of the DataFrame
print("-----------------------------all columns-------------------------------")
print(data)

# Display the last 5 rows of the Dataframe
print("-----------------------------Last 5 rows-------------------------------")
print(data.tail())

sns.set(style="whitegrid")

# Insight 1: Distribution of Genres
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Genre')
plt.title('Distribution of Video Game Genres')
plt.xticks(rotation=45)
plt.show()

# Insight 2: Global Sales over Years
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Year_of_Release', y='Global_Sales')
plt.title('Global Sales of Video Games over Years')
plt.show()

# Insight 3: Top Publishers by Global Sales
top_publishers = data.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_publishers.index, y=top_publishers.values)
plt.title('Top Publishers by Global Sales')
plt.xticks(rotation=45)
plt.show()

# Insight 4: User Score vs. Critic Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='User_Score', y='Critic_Score')
plt.title('User Score vs. Critic Score')
plt.show()

# Insight 5: Sales by Region for Top Games
top_games = data.nlargest(5, 'Global_Sales')
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
plt.figure(figsize=(10, 6))
sns.barplot(data=top_games, x='Name', y=regions)
plt.title('Sales by Region for Top Games')
plt.xticks(rotation=45)
plt.show()
```
